Financial Analytics Dashboards within Tableau. These calculations can help derive key metrics from the dataset consisting of customers, orders, location, and products.

### Calculated Fields for Sales Dashboard

1. **Total Sales**
   This formula calculates the total sales amount from the orders.
   ```
   SUM([Sales Amount])
   ```

2. **Total Profit**
   This formula calculates the total profit from the orders.
   ```
   SUM([Profit])
   ```

3. **Total Quantity Sold**
   This formula sums up the total quantity sold from the orders.
   ```
   SUM([Quantity])
   ```

4. **Year-over-Year Sales Growth**
   This formula calculates the percentage growth in sales compared to the previous year.
   ```
   (SUM(IF YEAR([Order Date]) = YEAR(TODAY()) THEN [Sales Amount] END) -
   SUM(IF YEAR([Order Date]) = YEAR(TODAY()) - 1 THEN [Sales Amount] END)) /
   SUM(IF YEAR([Order Date]) = YEAR(TODAY()) - 1 THEN [Sales Amount] END)
   ```

5. **Average Sales per Week**
   This formula calculates the average sales amount per week for the current year.
   ```
   SUM([Sales Amount]) / COUNTD(DATEPART('week', [Order Date]))
   ```

### Calculated Fields for Customer Dashboard

1. **Total Customers**
   This formula counts the distinct number of customers.
   ```
   COUNTD([Customer ID])
   ```

2. **Total Orders per Customer**
   This formula calculates the total number of orders placed per customer.
   ```
   COUNT([Order ID]) / COUNTD([Customer ID])
   ```

3. **Average Sales per Customer**
   This formula calculates the average sales amount for each customer.
   ```
   SUM([Sales Amount]) / COUNTD([Customer ID])
   ```

4. **Top Customers by Profit**
   This formula ranks customers based on their total profit.
   ```
   RANK(SUM([Profit]), 'desc')
   ```

5. **Customer Retention Rate**
   This formula calculates the retention rate by comparing the number of returning customers to the total number of customers.
   ```
   COUNTD(IF [Is Returning Customer] THEN [Customer ID] END) /
   COUNTD([Customer ID])
   ```

### Additional Notes

- Ensure that the field names (e.g., `[Sales Amount]`, `[Profit]`, `[Order Date]`, `[Customer ID]`) match the actual names used in your Tableau data source.
- These calculated fields can be used in various visualizations, helping to analyze sales performance and customer behavior effectively.
- You can customize the formulas as needed based on the specific requirements of your dashboards.
